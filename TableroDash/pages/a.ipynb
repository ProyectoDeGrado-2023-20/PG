```python
import pandas as pd
```

```python
df1 = pd.read_csv('../Data/Mapa_Distancia_IPS.csv')
df2 = pd.read_csv('../Data/Mapa_Distancia_IPS_Publicas.csv')
```

```python
df1.drop('Unnamed: 0', axis=1, inplace=True)
df2.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
total_pob = df1['Población'].sum()
```

```python
df1['Distancia_Poblacion_IPS_Nivel_1'] = df1['Distancia_IPS_Nivel_1'] * df1['Población'] / total_pob
df1['Distancia_Poblacion_IPS_Nivel_2'] = df1['Distancia_IPS_Nivel_2'] * df1['Población'] / total_pob
df1['Distancia_Poblacion_IPS_Nivel_3'] = df1['Distancia_IPS_Nivel_3'] * df1['Población'] / total_pob
```

```python
df2['Distancia_Poblacion_IPS_Nivel_1'] = df2['Distancia_IPS_Nivel_1'] * df2['Población'] / total_pob
df2['Distancia_Poblacion_IPS_Nivel_2'] = df2['Distancia_IPS_Nivel_2'] * df2['Población'] / total_pob
df2['Distancia_Poblacion_IPS_Nivel_3'] = df2['Distancia_IPS_Nivel_3'] * df2['Población'] / total_pob
```

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.merge(df1, df2, how='inner', left_index=True, right_index=True, suffixes=['_Modelo', '_Publicas'])
```

```python
df.columns
```

```python
df.head()
```

```python
df = df[['Codigo_Departamento_Modelo', 'Departamento_Modelo',
       'Codigo_Municipio_Modelo', 'Municipio_x_Modelo',
       'Municipio_Departamento_Modelo', 'Población_Modelo',
       'Distancia_IPS_Nivel_1_Modelo', 'Distancia_IPS_Nivel_2_Modelo',
       'Distancia_IPS_Nivel_3_Modelo',
       'Distancia_Poblacion_IPS_Nivel_1_Modelo',
       'Distancia_Poblacion_IPS_Nivel_2_Modelo',
       'Distancia_Poblacion_IPS_Nivel_3_Modelo',
       'Distancia_IPS_Nivel_1_Publicas', 'Distancia_IPS_Nivel_2_Publicas',
       'Distancia_IPS_Nivel_3_Publicas',
       'Distancia_Poblacion_IPS_Nivel_1_Publicas',
       'Distancia_Poblacion_IPS_Nivel_2_Publicas',
       'Distancia_Poblacion_IPS_Nivel_3_Publicas']]
```

```python
df
```

```python
df.rename({
    'Codigo_Departamento_Modelo': 'Codigo_Departamento', 
    'Departamento_Modelo': 'Departamento',
    'Codigo_Municipio_Modelo': 'Codigo_Municipio',
    'Municipio_x_Modelo': 'Municipio',
    'Municipio_Departamento_Modelo': 'Municipio_Departamento',
    'Población_Modelo': 'Población'}, axis=1, inplace=True)
```

```python
df
```

```python
df.to_csv('../Data/Mapa_Distancia_IPS.csv', index=False)
```

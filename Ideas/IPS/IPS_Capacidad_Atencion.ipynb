# Transformacion

```python
df_atencion_capacidad_transformed = df_atencion_capacidad.pivot_table(
    values='num cantidad capacidad instalada',
    # index=['codigo_sede', 'num nivel atencion', 'naturaleza'],
    index=['codigo_sede'],
    columns='nom descripcion capacidad ',
    aggfunc='sum',
    # margins=True
).fillna(0).reset_index()
df_atencion_capacidad_transformed = pd.merge(df_atencion_capacidad_transformed, df_atencion_capacidad[['codigo_sede', 'num nivel atencion', 'naturaleza']].drop_duplicates(), how='left', on='codigo_sede')
df_atencion_capacidad_transformed = df_atencion_capacidad_transformed[['codigo_sede', 'num nivel atencion', 'naturaleza'] + list(df_atencion_capacidad_transformed.columns)[1:-2]]
df_atencion_capacidad_transformed[list(df_atencion_capacidad_transformed.columns)[3:]] = df_atencion_capacidad_transformed[list(df_atencion_capacidad_transformed.columns)[3:]].astype('int64')
```

# Distancias

```python
mun_codigos = {}
```

```python
for i in range(0, len(df_municipios_ubicacion)):
    print(i)
    mun = df_municipios_ubicacion.iloc[i]
    mun_codigos[mun['Código_Mun']] = {
        'Nombre': mun['Municipio'],
        'Latitud': mun['Latitud'],
        'Longitud': mun['Longitud'],
    }
    break
```

```python
def calcular_distancia(coord1, coord2):
    return geodesic(coord1, coord2).kilometers
```

```python
coord1 = (6.257590, -75.611031)
coord2 = (5.803728, -75.438474)
# coord2 = (6.365534,	-75.090597)
```

```python
df_distancias = pd.DataFrame(index=df_municipios_ubicacion.index, columns=df_municipios_ubicacion.index)
```

```python
df_distancias
```

```python
columnas = df_municipios_ubicacion.index
```

```python
for i in (df_distancias):
    codigo_mun_i = i
    mun_i = df_municipios_ubicacion.loc[codigo_mun_i]
    lat_long_i = mun_i['Latitud'], mun_i['Longitud']
    
    for j in df_distancias.columns:
        if i == j:
            distancia_ij = 0
            df_distancias.at[i, j] = distancia_ij
            
        else:
            codigo_mun_j = j
            mun_j = df_municipios_ubicacion.loc[codigo_mun_j]
            lat_long_j = mun_j['Latitud'], mun_j['Longitud']            
            distancia_ij = calcular_distancia(lat_long_i, lat_long_j)            
            df_distancias.at[i, j] = distancia_ij
            # break
    # break
```
